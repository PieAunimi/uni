```python
#1.1importo le librerie 
import numpy as np 
import pandas as pd 
```

```python
#importo il csv
data=pd.read_csv('Animali_Urbani.csv',sep=';')
```

```python
#1.2determino il dataframe
animali=pd.DataFrame(data,columns=["Borough","Species Description","Species Status","Animal Condition","Duration of Response","Age","Animal Class"])
```

```python
#1.3rinnomino le colonne 
animali=animali.rename(columns={"Borough":"Quartiere di NY",
                                "Species Description":"Specie dell'animale",
                                "Species Status":"Specie domestica o selvatica",
                                "Animal Condition":"Stato di salute dell'animale",
                                "Duration of Response":"Durata dell'intervento",
                                "Age":"Fascia di età dell'animale",
                                "Animal Class":"Tipologia generale"})
```
