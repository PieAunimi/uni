```python
import numpy as np
import pandas as pd 
```

```python
dati=pd.read_csv('president_heights.csv')
```

```python
#definire le colonne ATTENTO AGLI SPAZI PD!
dataframe=pd.DataFrame(dati,columns=['order','name','height(cm)'])
print(dataframe)
```

```python

rslt_df=dataframe.loc[dataframe['height(cm)']<180]
print("\n altezza presidenti:\n",rslt_df)
```

```python
#determinare il dataframe poi aggiungere loc con nome colonna designata e limite richiesto 
#se vi sono due limiti aggiungerli dopo dando il nome del dataframe precedente mandare il print dellultimo dataframe 
rslt_df=dataframe.loc[dataframe['height(cm)']>=170]
rslt1_df=rslt_df.loc[rslt_df['height(cm)']<=180]
print("\n altezza presidenti:\n",rslt1_df)
```
